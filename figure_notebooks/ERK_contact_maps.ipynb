```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import seaborn as sns
sns.set_style('ticks')
sns.set_context('paper')
```

```python
erk2_wt = np.load('contact_maps/contact_map_erk2_wt.npy')
erk2_55n = np.load('contact_maps/contact_map_erk2_55n.npy')
erk2_q102m = np.load('contact_maps/contact_map_erk2_q102m.npy')
erk2_55n_q102m = np.load('contact_maps/contact_map_erk2_55n_q102m.npy')
```

```python
erk2_wt.shape
```

```python
erk2_55n.shape
```

```python
erk2_55n_wtsize = np.zeros((357,357))

for i in range(357):
    for j in range(357):
        if i >= 54:
            i_ = i+1
        else:
            i_ = i
        if j >= 54:
            j_ = j+1
        else:
            j_ = j
        erk2_55n_wtsize[i,j] = erk2_55n[i_,j_]    
```

```python
erk2_55n_q102m_wtsize = np.zeros((357,357))

for i in range(357):
    for j in range(357):
        if i >= 54:
            i_ = i+1
        else:
            i_ = i
        if j >= 54:
            j_ = j+1
        else:
            j_ = j
        erk2_55n_q102m_wtsize[i,j] = erk2_55n_q102m[i_,j_]    
```

```python
# mutant is at 0-index 54
```

```python
sns.set(font_scale=2.5, rc={'figure.figsize':(11.7*1.5,8.27*1.5)})
sns.set_style('ticks')
```

```python
plt.figure(dpi=300)
plt.imshow(erk2_wt, cmap='Greys')
cbar = plt.colorbar()
cbar.set_label('Mean contact fraction')
plt.title('ERK2 WT')
plt.xticks(np.arange(0,erk2_wt.shape[0],50))
plt.yticks(np.arange(0,erk2_wt.shape[0],50))
plt.show()
```

```python
contact_map = erk2_55n_wtsize - erk2_wt

contact_map_ = np.zeros(contact_map.shape)

for i, x in enumerate(contact_map):
    for j, y in enumerate(x):
        if j > i:
            break
        else:
            contact_map_[i,j] = contact_map[i,j]

# if np.max(contact_map_) > np.absolute(np.min(contact_map_)):
#     vmax = np.max(contact_map_)
#     vmin = - vmax
# else:
#     vmin = np.min(contact_map_)
#     vmax = -vmin      

vmin = -1
vmax = 1

plt.figure(dpi=300)
plt.imshow(erk2_wt, cmap='Greys')
plt.imshow(contact_map_, cmap='seismic_r', vmin=vmin, vmax=vmax, alpha=0.85)
plt.xticks(np.arange(0,contact_map.shape[0],50))
plt.yticks(np.arange(0,contact_map.shape[0],50))
cbar = plt.colorbar()
cbar.set_label('Mean contact fraction change')
plt.title('ERK2 55N - WT')
plt.show()
```
